```python
import serial
ard = serial.Serial("COM9", 9600, timeout=0.1)
```

```python
reeks = [bytearray("M 0 100 \r", "utf8" ),
         bytearray("M 1 45 \r", "utf8" ),
         bytearray("M 0 -100 \r", "utf8" ),
         bytearray("M 1 -45 \r", "utf8" ),
         bytearray("M 2 20 \r", "utf8" ),
         bytearray("M 3 20 \r", "utf8" ),
         bytearray("M 2 -20 \r", "utf8" ),
         bytearray("M 3 -20 \r", "utf8" )
        ]
```

```python
import time

i = 0

ard.flushInput()

while i < len(reeks):
    ard.write(reeks[i])
    
    time.sleep(0.1)
    antwoord = ard.readline()
    print(antwoord)
    
    while antwoord != b'Ready for next command\r\n':
        
        ard.flushInput()
        ard.write("Ready \r".encode())
        
        antwoord = ard.readline()
        print(antwoord)

            
    i+=1
```

```python
from tkinter import *
import time


root = Tk()

info = Label(root, text="[command] [step index] [distance / degrees]")
info.pack()

comando1 = Label(root, text="Comando 1:")
comando1.pack()
comando1_input = Entry(root, width=50)
comando1_input.pack()


comando2 = Label(root, text="Comando 2:")
comando2.pack()
comando2_input = Entry(root, width=50)
comando2_input.pack()


comando3 = Label(root, text="Comando 3:")
comando3.pack()
comando3_input = Entry(root, width=50)
comando3_input.pack()





def myClick():
    i = 0
    reeks = [bytearray(comando1_input.get()+"\r", "utf8" ),
             bytearray(comando2_input.get()+"\r", "utf8" ),
             bytearray(comando3_input.get()+"\r", "utf8" )
            ]
    

    ard.flushInput()

    while i < len(reeks):
        ard.write(reeks[i])
        ard.write(reeks[i+1])
        ard.write(reeks[i+2])
    
        time.sleep(1)
        antwoord = ard.readline()
        print(antwoord)
    
        while antwoord != b'Ready for next command\r\n':
        
            ard.flushInput()
            ard.write("Ready \r".encode())
        
            antwoord = ard.readline()
            print(antwoord)

            
        i+=3



myButtonLeft = Button(root, text="Send command", command=myClick)
myButtonLeft.pack()


root.mainloop()
```
